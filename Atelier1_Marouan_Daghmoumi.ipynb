*   Get all title of News that are in the link :

```python
def time():
  page = requests.get("https://www.bbc.com/arabic")
  src = page.content
  soup = BeautifulSoup(src, "lxml")
  times = soup.find_all("time", {'class': 'promo-timestamp bbc-1chh04s e1mklfmt0'})
  for time in times:
        news_time = time.text.strip()
        print(news_time)
```

*   Get content that are in the link that exest in the title HTML

```python
def get_content(link):
    try:
        # Send a GET request to the link
        response = requests.get(link)
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find all <p> elements with the specified class for content
        paragraphs = soup.find_all('p', class_='bbc-1gjryo4 e17g058b0')
        # Concatenate the text of all paragraphs
        content = ' '.join(paragraph.text.strip() for paragraph in paragraphs)
        return content
    except Exception as e:
        print(f"Error accessing {link}: {str(e)}")
        return None
```

*   Exemple

```python
get_content("https://www.bbc.com/arabic/articles/ckk7kq4r1e5o")
```

*   Get Writer that are in the link that exest in the title HTML

```python
def get_writer(link):
    try:
        # Send a GET request to the link
        response = requests.get(link)
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find the writer info
        writer = soup.find('span', class_='bbc-9scsxr')
        return writer.text.strip() if writer else None
    except Exception as e:
        print(f"Error accessing {link}: {str(e)}")
        return None
```

*   Exemple

```python
get_writer("https://www.bbc.com/arabic/articles/ckk7kq4r1e5o")
```


# **Visualize all**

```python
def visualize():
    try:
        # Send a GET request to the main page
        response = requests.get("https://www.bbc.com/arabic")
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find all links with the specified class
        links = soup.find_all('a', class_='focusIndicatorDisplayBlock bbc-uk8dsi e1d658bg0')
        # Find all times with the specified class
        times = soup.find_all("time", {'class': 'promo-timestamp bbc-1chh04s e1mklfmt0'})

        # Determine the minimum length between links and times
        min_length = min(len(links), len(times))

        articles = []
        # Extract the href attribute from each link and process them
        for i in range(min_length):
            # Get title from the current link
            title = links[i].text.strip()
            # Get time from the corresponding time element
            time = times[i].text.strip()
            href = links[i].get('href')
            # Get content and writer
            content = get_content(href)
            writer = get_writer(href)
            # Append article to list
            articles.append({
                'title': title,
                'time': time,
                'content': content,
                'writer': writer
            })
        return articles
    except Exception as e:
        print(f"Error: {str(e)}")
```

```python
articles = visualize()
visualize()
```

# **Save To MongoDB**

```python
def save_to_mongodb(articles):
    try:
        # Connect to MongoDB
        client = MongoClient('localhost', 27017)
        # Create or access the "Atelier1" database
        db = client['Atelier1']
        # Access the collection where the data will be stored
        collection = db['News_data']
        # Insert articles into MongoDB, except the last one
        for  i,article in articles[:-1]:
            collection.insert_one(article)
            print(f"Article {i+1} inserted into MongoDB.")
    except Exception as e:
        print(f"Error: {str(e)}")
```

```python
save_to_mongodb(articles)
```
